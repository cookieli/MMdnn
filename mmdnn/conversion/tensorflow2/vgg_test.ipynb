```python
def is_two_json_same(jsonfile1, jsonfile2):
    import json
    file1 = open(jsonfile1)
    file2 = open(jsonfile2)
    obj1 = json.load(file1)
    obj2 = json.load(file2)

    def ordered(obj):
        if isinstance(obj, dict):
            return sorted((k, ordered(v)) for k, v in obj.items() if k != 'op')
        if isinstance(obj, list):
            return sorted(ordered(x) for x in obj)
        else:
            return obj
    return ordered(obj1) == ordered(obj2)
print(is_two_json_same("tmp.json", "tmp2.json"))
```
